```python
# Project is based partially on data from MetroWind's project. However the code itself is made from scrath.
# Author: Si_ja
# Date: 2019/07/06
# References: Characters for generation are used from MetroWind's project
#             > https://gist.github.com/MetroWind/1401473/4631da7a4540a63e72701792a4aa0262acc7d397




def full_zalgo(text, up = True, mid = True, down = True, level = 2, gaps = False, gradual = False):
    """Ending function.
    All descriptions are further"""
    import sys, os, io, random
    import numpy as np

    # Characters taken from http://str.blogsite.org/Zalgo.htm.
    # And for me - taken from MetroWind/zalgo.py (on GitHub) as a reference information
    CHAR_UP = ['\u030D', '\u030E', '\u0304', '\u0305', '\u033F', '\u0311', '\u0306', '\u0310', '\u0352', '\u0357',
               '\u0351', '\u0307', '\u0308', '\u030A', '\u0342', '\u0343', '\u0344', '\u034A', '\u034B', '\u034C',
               '\u0303', '\u0302', '\u030C', '\u0350', '\u0300', '\u0301', '\u030B', '\u030F', '\u0312', '\u0313',
               '\u0314', '\u033D', '\u0309', '\u0363', '\u0364', '\u0365', '\u0366', '\u0367', '\u0368', '\u0369',
               '\u036A', '\u036B', '\u036C', '\u036D', '\u036E', '\u036F', '\u033E', '\u035B', '\u0346', '\u031A']

    CHAR_MID = ['\u0315', '\u031B', '\u0340', '\u0341', '\u0358', '\u0321', '\u0322', '\u0327', '\u0328', '\u0334',
                '\u0335', '\u0336', '\u034F', '\u035C', '\u035D', '\u035E', '\u035F', '\u0360', '\u0362', '\u0338',
                '\u0337', '\u0361', '\u0489']

    CHAR_DOWN = ['\u0316', '\u0317', '\u0318', '\u0319', '\u031C', '\u031D', '\u031E', '\u031F', '\u0320', '\u0324',
                 '\u0325', '\u0326', '\u0329', '\u032A', '\u032B', '\u032C', '\u032D', '\u032E', '\u032F', '\u0330',
                 '\u0331', '\u0332', '\u0333', '\u0339', '\u033A', '\u033B', '\u033C', '\u0345', '\u0347', '\u0348',
                 '\u0349', '\u034D', '\u034E', '\u0353', '\u0354', '\u0355', '\u0356', '\u0359', '\u035A', '\u0323']

    def zalgo(text, up = up, mid = mid, down = down, level = level, gaps = gaps, gradual = gradual):
        """This will make a zalgo text. Several conditions for it:
        > Text - actual text you want to scramble and add that zalgo effect.
        > Up - if True, text will be scrambled going up. By default is true.
        > Mid - if True, text be scrambled in the middle. By default is true.
        > Down - if True, text will be scrambled going down. By default is true.
        > Level - how severelly the text is scrambled. There are only 3 levels.
        > Gaps - will make gaps in your text if you want. By default it is off.
        > Gradual - will make the text gradually go insane. By default it is off.

        Inside there is also another function that is nested. It's job is to scramble the text.
        Its presence will save on space that we need to utilize in our code for making of the scrambler,
        because there are up to 3 procedures that need to be performed otherwise."""

        #First indicate the scrambler procedure
        def scrambler(text, direction = None, level = None, gaps = False, gradual = False):
            """Scrambler will add that zalgo effect to our code. Conditions will be passed automatically.
            > Text - from before used text.
            > Direction - to what direction the zalgoing will happe, up, midway or down.
            > Level - how severelly it will be scrambled.
            > Gaps - from before will intentionllay make gaps in your text. By default it is off. Chance: 1/2 of a gap."""

            #Scrambling pre-defined levels to up to how many values could be scrambled from lower bounds to upper bounds
            low_l = 1 
            low_u = 3

            med_l = 3 
            med_u = 6

            hig_l = 6 
            hig_u = 10

            #Prepare our scrambler        
            if level == 1:
                lower = low_l 
                upper = low_u
            elif level == 2:
                lower = med_l 
                upper = med_u
            else:
                lower = hig_l 
                upper = hig_u

            #Decide from which set we will be zalgoing our letters:
            if direction == "up":
                ref_set = CHAR_UP
            elif direction == "mid":
                ref_set = CHAR_MID
            else:
                ref_set = CHAR_DOWN

            #Now we need to know the boundries of our sets
            set_l = 0
            set_u = len(ref_set)

            #Now we can ZALGO IT!
            #Step 1: Break text into characters:
            new_text = list(text)

            #Step 2: For each letter, whitespace or any character there is, apply a scrambler:
            #A lot of randomization occurs here

            if gradual is True:
                gradual_eval = len(text)

            for position, character in enumerate(new_text):
                if gradual is True:
                    if gradual_eval*0.15 >= position:
                        continue
                    elif position <= gradual_eval/2:
                        if np.random.randint(0, 101) <= 20:
                            continue
                #Amount of scrambles that will apply to each letter
                i = np.random.randint(lower, upper + 1) #End point
                i_i = 0                                 #Start point

                while i_i <= i:

                    #Decide which letter will character will be added as scrambled
                    val = np.random.randint(set_l, set_u)
                    if gaps == True and np.random.randint(0, 101) <= 50:
                        i_i += 1
                        continue
                    #Add the scrambler
                    character = character + ref_set[val]
                    #Increase the itterator
                    i_i += 1

                new_text[position] = character

            #Step 3: put the text back together:

            final_text = "".join(new_text)

            #Step 4: return the finalized updated text

            return final_text

        #This part adjusts the level value if one is indicated incorrectly
        #More of a breaker check than anything
        if level not in [1, 2, 3]:
            if level <= 0:
                level = 1
            elif level >= 4:
                level = 3
            else:
                level = 2

        #Assess where our values will have to be applied (up, mid, down):
        list_changes = []
        if up is True:
            list_changes.append("up")
        if mid is True:
            list_changes.append("mid")
        if down is True:
            list_changes.append("down")

        #and lets scramble our text for as much as possible...but first we will also shuffle the list for more randomness:
        random.shuffle(list_changes)
        for insanity in list_changes:
            text = scrambler(text = text, direction = str(insanity), level = level, gaps = gaps, gradual = gradual)

        #Save the text:
        file = io.open("zalgo.txt", "w", encoding="utf-8")
        file.write("\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n" + text) 
        file.close()

        return text
    
    zalgo(text, up = up, mid = mid, down = down, level = level, gaps = gaps, gradual = gradual)
```
